```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://en.wikipedia.org/wiki/List_of_hospitals_in_India"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.prettify())
```

```python
list_of_hospitals = []

for x in soup.find_all("li")[113:]:
    if x=="Healthcare in India":
        break
    else:
        list_of_hospitals.append(x.get_text())
```

```python
list_of_hospitals = list_of_hospitals[:357]
```

```python
pd.DataFrame(list_of_hospitals, columns=["hospital_name"]).to_csv("hospitals.csv", index=False)
```
